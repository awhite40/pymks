```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
    
```

```python
import pymks

L= 21
center = (L-1)/2
n_samples = 364
#elastic_modulus = ()
#poissons_ratio = 
```

```python
# import microstructures
import scipy.io

mat = scipy.io.loadmat('Elong_all.mat')
El = mat['ElongFib']
e11 = mat['e11']
#El = El[0:2]
#e11 = e11[0:2]
El.shape
```

```python
from pymks.tools import draw_microstructure_strain

draw_microstructure_strain(El[1, center, :, :], e11[1, center, :, :])
```

```python
import sklearn
from sklearn.cross_validation import train_test_split
test_size = 0.5
split_shape = (El.shape[0],) + (np.product(El.shape[1:]),)
El_train, El_test, e11_train, e11_test = train_test_split(El.reshape(split_shape),
                                                          e11.reshape(split_shape),
                                     test_size=test_size, random_state=3)

El_train = El_train.reshape(n_samples*(1-test_size),21,21,21)
El_train.shape
e11_train=e11_train.reshape(n_samples*(1-test_size),21,21,21)
El_test = El_test.reshape(n_samples*(test_size),21,21,21)
e11_test=e11_test.reshape(n_samples*(test_size),21,21,21)
draw_microstructure_strain(El_train[0, center, :, :], e11_train[0, center, :, :])
```

```python
# Calibrate influence coefficients with 2 local states
from sklearn import metrics
mse = metrics.mean_squared_error
from pymks import MKSRegressionModel
from pymks.bases import DiscreteIndicatorBasis


basis=DiscreteIndicatorBasis(n_states=2)
MKSmodel = MKSRegressionModel(basis=basis)
MKSmodel.fit(El_train,e11_train)
```

```python
from pymks.tools import draw_coeff

coeff = MKSmodel.coeff
draw_coeff(coeff[ center])
```

```python
# Predict the strain field
from pymks.tools import draw_microstructure_strain
#draw_microstructure_strain(El_test[1,10], e11_test[1,10])
```

```python
# Compare strain field with one calculated from Abaqus

from pymks.tools import draw_strains_compare


strain_pred = MKSmodel.predict(El_test)
draw_strains_compare(e11_test[0, center], strain_pred[0, center])
```
